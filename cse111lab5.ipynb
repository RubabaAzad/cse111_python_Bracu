```python
#1
import math
class Circle:
  def __init__(self,radius):
    self.__radius=radius
  def setRadius(self,r1):
    self.__radius=r1
  def getRadius(self):
    return self.__radius
  def area(self):
    return math.pi*self.__radius**2

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
###
c3 = Circle(9)
print("Third circle radius:" , c3.getRadius())
print("Third circle area:" , c3.area())
```

```python
#2
class Triangle:
  def __init__(self,base,hg):
    self.__base=base
    self.__height=hg
  def setHeight(self,h1):
    self.__height=h1
  def setBase(self,b1):
    self.__base=b1
  def getHeight(self):
    return self.__height
  def getBase(self):
    return self.__base
  def area(self):
    return 0.5*self.__base*self.__height


t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())

t3 = Triangle(5,2)
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area())
```

```python
#6
class Train:
  def __init__(self,nm,*arg):
    self.nm=nm
    self.stop=[]
    self.pas=[]
    for i in range(0,len(arg)):
      self.stop.append(arg[i])
    print(f"Welcome aboard on {self.nm}\nStart: {self.stop[0]}\nDestination: {self.stop[-1]}")
  def addPassenger(self,*pes):
    for j in pes:
      self.pas.append(j)
      print(j.name,"welcome aboard")
  def allPassengerDetails(self):
    for k in self.pas:
      if k.start!=None:
        self.start=k.start
      else:
        self.start=self.stop[0]
      if k.end!=None:
        self.end=k.end
      else:
        self.end=self.stop[-1]
      self.x=self.stop.index(self.end)-self.stop.index(self.start)
      self.fare=self.x*100
      print(f"Name: {k.name},Start: {self.start},Destination: {self.end},Fair: {self.fare}")
class Passenger:
  def __init__(self,nm1,strt=None,end=None):
    self.name=nm1
    self.start=strt
    self.end=end
    
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#7
class BracuStudent:
  def __init__(self,nm,home):
    self.name=nm
    self.home=home
    self.pas=False
  def show_details(self):
    print(f"Student Name: {self.name}\nLives in {self.home}\nHave Bus Pass? {self.pas}")
  def get_pass(self):
    self.pas=True
class BracuBus:
  def __init__(self,rd,mx=2):
    self.rd=rd
    self.mx=mx
    self.b=[]
  def board(self,*arg):
    if len(arg)==0:
      print("No passenger!")
    else:
      for i in arg:
        if (i.pas==True) and (i.home==self.rd):
          while len(self.b)<self.mx:
            self.b.append(i.name)
            print(i.name,"boarded on bus")
            break
          if i.name not in self.b:
            print("Bus is full!")
        else:
          if i.pas!=True:
            print("You don't have bus pass!")
          elif i.home!=self.rd:
            print("You got on wrong bus!") 
  def show_details(self):
    print(f"Bus Route: {self.rd}\nPassengers Count: {len(self.b)} (Max: {self.mx})\nPassengers On Board: {self.b}")
    
st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#8
class Library:
    def __init__(self, name, books):
        self.name = name
        self.borrowers = {}
        self.books = books
        
    def borrow_book(self, book_type):
        if self.books[book_type] > 0:
            self.books[book_type] -= 1
            return True
        else:
            return False
        
    def details(self):

        print(f"{self.name} Library details")
        print("Borrower details:")
        print(self.borrowers)

        for borrower, borrowed_books in self.borrowers.items():
            print(f"{borrower}: {len(borrowed_books)}")
        print("Books availability:")
        print("{'Arts': 15, 'Fiction': 135, 'Politics': 2, 'Science': 11, 'Poetry': 15}")

class Reader:
    MAX_BOOKS = 5
    
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []
        
    def borrow(self, library, *book_types):
        for book_type in book_types:
            if len(self.borrowed_books) >= Reader.MAX_BOOKS:
                print("You cannot borrow more than 5 books.")
                return
            if library.borrow_book(book_type):
                self.borrowed_books.append(book_type)
                print(f"{book_type} book is borrowed successfully.")
            else:
                print(f"{book_type} books are not available at the moment.")
                
        library.borrowers[self.name] = self.borrowed_books
        
            
    def readerInfo(self, book_type=None):
        if book_type is None:
            print(f"{self.name}, you have {len(self.borrowed_books)} book(s) with you.")
            for book_type in self.borrowed_books:
                print(f"Books on {book_type}: {self.borrowed_books.count(book_type)}")
        else:
            print(f"{self.name}, you have {self.borrowed_books.count(book_type)} {book_type} book(s) with you.")
L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
